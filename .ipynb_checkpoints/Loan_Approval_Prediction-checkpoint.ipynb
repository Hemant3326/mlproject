```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,precision_score,recall_score,f1_score,mean_squared_error,mean_absolute_error,r2_score
from sklearn.linear_model import Ridge, Lasso

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("LoanApprovalPrediction.csv") # Loading the data
```

```python
df.head()
```

```python
# Data pre-procesing 
```

```python
df.drop(["Loan_ID"], axis=1, inplace=True)
```

```python
df["Gender"].value_counts().to_dict()
```

```python
df["Gender"].replace({'Male': 1, 'Female': 0}, inplace=True) # Encoding Male as 1 & Female as 0 
```

```python
df["Gender"].value_counts().to_dict()
```

```python
df["Married"].value_counts().to_dict()
```

```python
df["Married"].replace({'Yes': 1, 'No': 0}, inplace=True) # Encoding Yes as 1 & No as 0 
```

```python
df["Married"].value_counts().to_dict()
```

```python
df["Education"].value_counts().to_dict()
```

```python
df["Education"].replace({'Graduate': 1, 'Not Graduate': 0}, inplace=True) # Encoding Graduate as 1 & Not Graduate as 0 
```

```python
df["Education"].value_counts().to_dict()
```

```python
df["Self_Employed"].value_counts().to_dict()
```

```python
df["Self_Employed"].replace({'No': 0, 'Yes': 1}, inplace = True) # Encoding Yes as 1 & Not No as 0 
```

```python
df["Self_Employed"].value_counts().to_dict()
```

```python
df["Loan_Status"].value_counts().to_dict()
```

```python
df["Loan_Status"].replace({'N': 0, 'Y': 1}, inplace = True) # Encoding Yes as 1 & Not No as 0 
```

```python
df["Loan_Status"].value_counts().to_dict()
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=["Property_Area"])
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
for col in df.columns:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot=True, cmap="Blues", annot_kws={"size": 7})
plt.figure(figsize=(25,25))
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(["Loan_Status"],axis = 1)
y = df["Loan_Status"]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
```

```python
#Logistic Regression
lr_model = LogisticRegression()
lr_model.fit(x_train,y_train)
```

```python
y_pred = lr_model.predict(x_test)

cnf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix \n", cnf_matrix)

acc = accuracy_score(y_test, y_pred)
print("Accuracy Score \n", acc)

clf_report = classification_report(y_test, y_pred)
print("Classificartion Report \n", clf_report)
```

```python
rfc_model = RandomForestClassifier(n_estimators = 7,
                             criterion = 'entropy',
                             random_state =7)
rfc_model.fit(x_train,y_train)
```

```python
y_pred_rfc = rfc_model.predict(x_test)

cnf_matrix_rfc = confusion_matrix(y_test, y_pred_rfc)
print("Confusion Matrix \n", cnf_matrix_rfc)

acc_rfc = accuracy_score(y_test, y_pred_rfc)
print("Accuracy Score \n", acc_rfc)

clf_report_rfc = classification_report(y_test, y_pred_rfc)
print("Classificartion Report \n", clf_report_rfc)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#For logistic regression
score = cross_val_score(lr_model, x, y, cv=5)
print("Cross validation is",np.mean(score)*100)
```

```python
#For random forest classifier
score_rfc = cross_val_score(rfc_model, x, y, cv=5)
print("Cross validation is",np.mean(score_rfc)*100)
```
